**Imports**

```python
# Essentials
import numpy as np
import pandas as pd 
import matplotlib.pyplot as plt
%matplotlib inline
import os

# Import functions for data preprocessing & data preparation
from sklearn.preprocessing import LabelEncoder
from sklearn.utils import resample
from sklearn.feature_extraction.text import CountVectorizer
from nltk.sentiment.vader import SentimentIntensityAnalyzer

from nltk.tokenize import word_tokenize
from nltk.stem import WordNetLemmatizer
from nltk.stem import PorterStemmer, LancasterStemmer
from nltk.stem.snowball import SnowballStemmer
from nltk.corpus import stopwords
from nltk.corpus import wordnet
import string
from string import punctuation
import nltk
import re
```

```python
data = pd.read_csv(f'comments-{VIDEO_ID}.csv')
data.columns
data1=data.drop(['kind', 'etag','id','snippet.channelId','snippet.videoId', '.kind',
       '.etag', '.id', 'channelId', 'videoId','textDisplay',
       'authorDisplayName', 'authorProfileImageUrl', 'authorChannelUrl',
       'authorChannelId.value', 'canRate', 'viewerRating', 'likeCount',
       'publishedAt', 'updatedAt', 'snippet.canReply',
       'snippet.totalReplyCount', 'snippet.isPublic', 'replies'],axis=1)
data1
```

**Data labelling**

```python
nltk.download('vader_lexicon')
sentiments = SentimentIntensityAnalyzer()
data1["Positive"] = [sentiments.polarity_scores(i)["pos"] for i in data1["textOriginal"]]
data1["Negative"] = [sentiments.polarity_scores(i)["neg"] for i in data1["textOriginal"]]
data1["Neutral"] = [sentiments.polarity_scores(i)["neu"] for i in data1["textOriginal"]]
data1['Compound'] = [sentiments.polarity_scores(i)["compound"] for i in data1["textOriginal"]]
score = data1["Compound"].values
sentiment = []
for i in score:
    if i >= 0.05 :
        sentiment.append('Positive')
    elif i <= -0.05 :
        sentiment.append('Negative')
    else:
        sentiment.append('Neutral')
data1["Sentiment"] = sentiment
data1.head()
```

**Finaldata**

```python
data2=data1.drop(['Positive','Negative','Neutral','Compound'],axis=1)
data2.head()
```

```python
stop_words = stopwords.words('english')
porter_stemmer = PorterStemmer()
lancaster_stemmer = LancasterStemmer() 
snowball_stemer = SnowballStemmer(language="english")
lzr = WordNetLemmatizer()
```

```python
def text_processing(text):   
    # convert text into lowercase
    text = text.lower()

    # remove new line characters in text
    text = re.sub(r'\n',' ', text)
    
    # remove punctuations from text
    text = re.sub('[%s]' % re.escape(punctuation), "", text)
    
    # remove references and hashtags from text
    text = re.sub("^a-zA-Z0-9$,.", "", text)
    
    # remove multiple spaces from text
    text = re.sub(r'\s+', ' ', text, flags=re.I)
    
    # remove special characters from text
    text = re.sub(r'\W', ' ', text)

    text = ' '.join([word for word in word_tokenize(text) if word not in stop_words])
    
    # stemming using porter stemmer from nltk package - msh a7sn 7aga - momken: lancaster, snowball
    # text=' '.join([porter_stemmer.stem(word) for word in word_tokenize(text)])
    # text=' '.join([lancaster_stemmer.stem(word) for word in word_tokenize(text)])
    # text=' '.join([snowball_stemer.stem(word) for word in word_tokenize(text)])
    
    # lemmatizer using WordNetLemmatizer from nltk package
    text=' '.join([lzr.lemmatize(word) for word in word_tokenize(text)])

    return text
```

```python
data_copy = data2.copy()
data_copy.textOriginal = data_copy.textOriginal.apply(lambda text: text_processing(text))
```

```python
le = LabelEncoder()
data_copy['Sentiment'] = le.fit_transform(data_copy['Sentiment'])
```

```python
processed_data = {
    'Sentence':data_copy.textOriginal,
    'Sentiment':data_copy['Sentiment']
}

processed_data = pd.DataFrame(processed_data)
processed_data.head()
```

```python
df_neutral = processed_data[(processed_data['Sentiment']==1)] 
df_negative = processed_data[(processed_data['Sentiment']==0)]
df_positive = processed_data[(processed_data['Sentiment']==2)]

# upsample minority classes
df_negative_upsampled = resample(df_negative, 
                                 replace=True,    
                                 n_samples= 205, 
                                 random_state=42)  

df_neutral_upsampled = resample(df_neutral, 
                                 replace=True,    
                                 n_samples= 205, 
                                 random_state=42)  


# Concatenate the upsampled dataframes with the neutral dataframe
final_data = pd.concat([df_negative_upsampled,df_neutral_upsampled,df_positive])
```

```python
final_data['Sentiment'].value_counts()
```

```python
corpus = []
for sentence in final_data['Sentence']:
    corpus.append(sentence)
corpus[0:5]
```

```python
from sklearn.feature_extraction.text import CountVectorizer
cv = CountVectorizer(max_features=1500)
X = cv.fit_transform(corpus).toarray()
y = final_data.iloc[:, -1].values
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=0)
```

```python
from sklearn.naive_bayes import GaussianNB
```

```python
classifier = GaussianNB()
classifier.fit(X_train, y_train)
```

```python
from sklearn.metrics import confusion_matrix, accuracy_score
y_pred = classifier.predict(X_test)
```

```python
cm = confusion_matrix(y_test, y_pred)
cm
```

```python
nb_score = accuracy_score(y_test, y_pred)*100
nb_score
```

```python
from sklearn.metrics import classification_report

print("Classification Report:\n", classification_report(y_test, y_pred))
```
